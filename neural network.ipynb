classification 하고자 하는 사람의 특성: 정보가 단순하고 알기 쉽기 때문에 discrete한 정보를 원한다(남자 아니면 여자 not 남자 78.8%, 여자 21.2%),사람은 주로 a이냐 아니냐,즉 non-linear하게 생각한다.

 -sigmoid : 0 또는 1로만 표시하게 함(어느쪽에 가까운지 classify)

regression의 끝부분에 sigmoid 걸어주면 classification이 가능하다.=non-linearity를 부여할 수 있다.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = pd.read_csv('regression.csv')   # pd: excel 파일 불러오는 프로그램(module)
print(data)
X = data['X']
Y = data['Y']
plt.plot(X, Y, '.')
```

```python
model = Sequential()  # model의 시작, 입력벡터에서 출력벡터로 가는 것이 하나의 sequence, 현재 1개의 시퀀스만 필요
model.add(Dense(1, input_shape=[1])) # y의 갯수 1로 설정(출력1개), input shape에서 x의 갯수 1로 설정(입력1개) - 입력차원 하나,출력차원 하나
model.compile(optimizer="sgd", loss='mse')  # model을 닫음, optimizer: 함수행렬 통해 도출된 예측 output과 실제 output을 비교해가며 차이를 조정해가는 역할: sgd는 그 차이 조정 방법 중 하나
#loss: 예측된 출력값(output)과 정답값(target)과의 차이를 어떻게 계산할 것인가.mse는 여러 계산 방법 중 하나(mse: mean squre --: 오차의 제곱의 평균)
```

```python
model = Sequential()
model.add(Dense(10, input_shape=[1]))
model.add(Dense(3))    # node가 3개 있는(graph에서 동그라미 3개 있는) hidden layer 삽입
model.add(Dense(6, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))  # hidden layer가 아니라 출력값 자리(1이 되어야 함) # activation 통해 마지막에 non-linearity 부여
#non-linearity는 layer별로 부여할 수 있음   # sigmoid 외에도 다양한 non-linearity부여 가능(tanh, relu 등)  #activation='linear':아무것도 안 적는 것과 동일
model.compile(optimizer='rmsprop', loss='mse')
```

# Practice

```python
data = pd.read_csv('regression.csv')
X = data['X']
Y = data['Y']
```

```python
model = Sequential()
model.add(Dense(5, input_shape=[1]))
model.add(Dense(3))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
```

```python
model.fit(X,Y, epochs=100)
```

```python
x = [1,3,5,7,9,11,13,15]
y = model.predict(x)
plt.plot(X, Y, '.')
plt.plot(x, y, 'x')
```
